```python
from tensorflow.keras.models import load_model
# Modeli yükle
model = load_model('modelv1.h5')
```

```python
import cv2
import numpy as np

# Webcam üzerinden video akışı al
cap = cv2.VideoCapture(0)

# Yüz tanıma için OpenCV'nin önceden eğitilmiş yüz tanıma modelini yükle
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Duygu etiketleri
emotion_labels = {0: 'angry', 1: 'happy'}

while True:
    ret, frame = cap.read()  # Kameradan bir frame al

    # Yüzleri algıla
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        face = gray[y:y+h, x:x+w]  # Yüz bölgesini gri tonlamalı olarak al
        face = cv2.resize(face, (48, 48))  # Model için giriş boyutuna yeniden boyutlandır
        face = np.expand_dims(face, axis=-1)  # Boyutları yeniden düzenle

        # Yüz ifadesi tahmini
        emotion_label = model.predict(np.array([face]))  # Modeli kullanarak tahmin yap

        # Tahmini etikete dönüştür
        emotion = emotion_labels[np.argmax(emotion_label)]

        # Sonucu ekrana yazdır
        cv2.putText(frame, emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        # Yüzü çevreleyen dikdörtgeni çiz
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Sonucu göster
    cv2.imshow('Video', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
